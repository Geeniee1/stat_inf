```python
import csv
from bs4 import BeautifulSoup

def parse_departures(html_path: str, line_number: str) -> list[dict]:
    with open(html_path, encoding="utf-8") as f:
        soup = BeautifulSoup(f, "lxml")

    results = []
    for item in soup.select("li.expandable-list-item"):
        line_span = item.select_one("span.fw-bold.small")
        if line_span is None or line_span.text.strip() != str(line_number):
            continue
        departure_span = item.select_one("span.visually-hidden:-soup-contains('departure time:')")
        if departure_span is None:
            continue
        time_span = departure_span.find_next_sibling("span")
        if time_span is None:
            continue
        results.append({"line": line_number, "departure_time": time_span.text.strip()})
    return results

def save_departures(html_path: str, line_number: str, output_csv: str) -> None:
    rows = parse_departures(html_path, line_number)
    with open(output_csv, "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=["line", "departure_time"])
        writer.writeheader()
        writer.writerows(rows)
    print(f"Saved {len(rows)} departures for line {line_number} → {output_csv}")
```

```python
# Line 6 — Kortedala via Centralstationen
save_departures(
    "data/raw_data/html_line6_direction_kortedala.txt",
    "6",
    "data/cleaned_data/line_6__kortedala_scheduled.csv",
)

# Line 6 — Länsmansgården via Sahlgrenska
save_departures(
    "data/raw_data/html_line6_direction_varmfront.txt",
    "6",
    "data/cleaned_data/line_6_varmfront_scheduled.csv",
)

# Line 64 — Heden
save_departures(
    "data/raw_data/html_line64_direction_heden.txt",
    "64",
    "data/cleaned_data/line_64_heden_scheduled.csv",
)

# Line 64 — Högsbohöjd
save_departures(
    "data/raw_data/line_64_direction_axel.txt",
    "64",
    "data/cleaned_data/line_64_axel_scheduled.csv",
)
```
